```python
from bs4 import BeautifulSoup
import requests
import re
import urllib
import pandas as pd
```

```python
df = pd.DataFrame(columns=["Job Id","Title","Location","Company","Salary","Job Post Link","Job Description"])

all_jobs = []
# salary = 0
# qjob = '/jobs?q='
# qstart = '&start=' + str(start)

for i in range(0, 10, 10):
    # page = requests.get('https://www.indeed.com/'+ qjob +qcountry+ str(start))
    page = requests.get('https://www.indeed.com/jobs?q=data+scientist&l=usa'+ '&start=' + str(i))
    soup = BeautifulSoup(page.text, 'lxml')
    job_posts = soup.find_all(class_= "jobsearch-SerpJobCard unifiedRow row result" )

    for post in job_posts:  
        job_id = post.get('id')
        if job_id in [d['jobId'] for d in all_jobs]:
            continue
        
        url = post.find('a', class_='jobtitle')
        href = url.get('href')
        title = post.find('a', class_='jobtitle').text.strip()
        company = post.find('span', class_='company').text.strip()
        
        try:
            location = post.find('span', class_='location').text.strip()
        except:
            try:
                location = post.find('div', class_= 'location').text.strip()
            except:
                location = post.find('span', class_='remote').text.strip()
                  
        
        try:
            salary = post.find('span', class_='no-wrap').text
        except:
            pass
            

        if None in (title, company, url,location):
            pass

        if href is None:
            pass
        
        job_link = f'https://www.indeed.com{href}'
        detail_page = requests.get(job_link)
        detail = BeautifulSoup(detail_page.text, 'lxml')
        
        job_description = detail.find('div', class_="jobsearch-jobDescriptionText")
        

        item = {
                    'jobId':job_id,
                    'title' : title,
                    'company' : company,
                    'href' : f'https://www.indeed.com{href}',
                    'salary' : salary,
                    'location':location,
                    'job description':job_description.text
                }
        all_jobs.append(item)
#         print(json.dumps(item))

        df = df.append({'Job Id':job_id,'Title':title, 'Location':location, 
                        'Company':company, 'Salary':salary.strip(),
                        'Job Post Link':job_link, 'Job Description': job_description.text}, ignore_index=True)
        
```
